## Gender Demographics - Done

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Steps
# 1 Create a separate df
# 2 Create a df showing all unique elements or drop duplicates
# 3 Get count of Male, female, others
# 4 Get count total
# 5 Compute for percentages
```

```python
# 1 Create a separate df
gender_demo_df = purchase_data[["Gender", "SN"]]
gender_demo_df
```

```python
# 2 Create a df showing all unique elements or drop duplicates
gender_demo_df.drop_duplicates(subset="SN", keep ="first", inplace=True)
gender_demo_df
```

```python
gender_count = gender_demo_df["SN"].count()
gender_count
```

```python
# 3 Get count of Male, female, others
gender_demo_count_df = gender_demo_df.groupby(["Gender"]).count()
gender_demo_count_df = gender_demo_count_df.rename(columns={"SN":"Total Count"})
gender_demo_count_df
```

```python
gender_demo_count_df["Percentage"] = ((gender_demo_count_df["Total Count"]/gender_count)*100).map("{:.2f}%".format)
gender_demo_count_df
```

```python
gender_demo_sorted_df = gender_demo_count_df.sort_values(["Total Count"], ascending=False)
gender_demo_sorted_df
```


## Purchasing Analysis (Gender) - Done

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Steps
# 1 Create dataframe
purchasing_analysis_data = purchase_data[["Gender", "Price", "SN"]]
purchasing_analysis_data
```

```python
purchasing_analysis_grpby = purchasing_analysis_data.groupby(['Gender'])
purchasing_analysis_grpby
```

```python
# 2 Compute for purchase count per gender type
gen_pur_ana_pur_count = purchasing_analysis_grpby['Gender'].count()
gen_pur_ana_pur_count
```

```python
# 3 Compute for average purchase price per gender type
gen_pur_ana_avg_price = purchasing_analysis_grpby['Price'].mean()
gen_pur_ana_avg_price
```

```python
# 4 Compute for total purchase value per gender type
gen_pur_ana_tot_pur_value = purchasing_analysis_grpby['Price'].sum()
gen_pur_ana_tot_pur_value
```

```python
# 5 Compute for Average Total Purchase per person
# 5a compute for unique purchasers through SN, classify per gender
gen_pur_ana_SN_uniq = purchasing_analysis_grpby['SN'].nunique()
gen_pur_ana_SN_uniq
```

```python
# 5b Divide Total Purchase Value by 5a result
gen_pur_ana_ave_tot_pur_person = gen_pur_ana_tot_pur_value/gen_pur_ana_SN_uniq
gen_pur_ana_ave_tot_pur_person
```

```python
# 6 Create Dataframe

purchasing_analysis_df = pd.DataFrame({"Purchase Count": gen_pur_ana_pur_count,
                                       "Average Purchase Price": gen_pur_ana_avg_price,
                                       "Total Purchase Value": gen_pur_ana_tot_pur_value,
                                       "Avg Total Purchase per Person": gen_pur_ana_ave_tot_pur_person,
                                        })

purchasing_analysis_df
```

```python
# 7 Duplicate and format DataFrame
purchasing_analysis_formatted_df = purchasing_analysis_df

purchasing_analysis_formatted_df["Average Purchase Price"] = purchasing_analysis_df["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_formatted_df["Total Purchase Value"] = purchasing_analysis_df["Total Purchase Value"].map("${:.2f}".format)
purchasing_analysis_formatted_df["Avg Total Purchase per Person"] = purchasing_analysis_df["Avg Total Purchase per Person"].map("${:.2f}".format)
purchasing_analysis_formatted_df
```

## Age Demographics - Done

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Steps
# 1 Create source data
# 2 Drop duplicates
# 3 Create bins
# 4 Create names for the bins
# 6 Create a GroupBy object based upon "Age Group"
# 7 Count the frequency per "Age Group"
# 8 Compute for sum of total count
# 9 Compute for percentage of players
```

```python
# 1 Create source data
age_demo_data = purchase_data[["Age", "SN"]]
age_demo_data
```

```python
# 2 Drop duplicates
age_demo_data.drop_duplicates(subset="SN", keep ="first", inplace=True)
age_demo_data
```

```python
# 3 Create bins
bins = [0,9,14,19,24,29,34,39,100]

# 4 Create names for the bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
# 5 Place the data series into a new column inside of the DataFrame
age_demo_data["Age Group"] = pd.cut(age_demo_data["Age"], bins, labels=group_names, include_lowest=True)
age_demo_data
```

```python
# 6 Create a GroupBy object based upon "Age Group"
age_demo_df = age_demo_data.groupby(["Age Group"])

# 7 Count the frequency per "Age Group"
age_demo_df2 = age_demo_df[["SN"]].count()
age_demo_df2 = age_demo_df2.rename(columns={"SN":"Total Count"})
age_demo_df2
```

```python
# 7 Compute for sum of total count
age_count = age_demo_df2["Total Count"].sum()
age_count
```

```python
# 8 Compute for percentage of players
age_demo_df2["Percentage of Players"]=((age_demo_df2["Total Count"]/age_count)*100).map("{:.2f}%".format)
age_demo_df2
```

## Most Profitable Items - Done

```python
# Slice source data from main data
profit_items_data = purchase_data[["Item ID", "Item Name", "Purchase ID", "Price"]]
profit_items_data
```

```python
# Groupby
pro_ite_grpby = profit_items_data.groupby(["Item ID", "Item Name"])
pro_ite_grpby
```
